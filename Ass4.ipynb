```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean=100
standard_deviation=4
dataset_size=100000
```

```python
data= np.random.normal(loc=mean, scale=standard_deviation, size=dataset_size)
```

```python
#Visualization of the normal distribution using histogram
plt.figure(figsize=(10,6))
plt.hist(data, bins=100, color='grey')
plt.title(f'Normal Distribution (mean = {mean}, std_dev = {standard_deviation}, dataset size = {dataset_size})')
plt.show()
```

```python
# This graph shows the bell curved shape.most data points are around the mean.
# The spread of the data is controlled by the standard deviation.
# About 68% of the data falls within one standard deviation,and 95% within two standard deviations.
```
